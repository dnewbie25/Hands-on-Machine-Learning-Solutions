### Regresion Lineal

##### **Pregunta:** Al generar un modelo de regresión lineal simple para cada variable por separado, es decir, y1=f(x1), y2=f(x2) y y3= f(x3), con y1, y2, y3 la variable 'Sales', utilizando el 100% del dataset y trabajando con los hiperparámetros por defecto, la mayor bondad de ajuste se logra para...

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Advertising.csv")
data.head()
```

```python
feature_cols = ["TV", "Radio", "Newspaper"]
X = data[feature_cols]
y = data["Sales"]
x1 = X[["TV"]]
x2 = X[["Radio"]]
x3 = X[["Newspaper"]]
```

```python
modelo_rl1 = LinearRegression()
modelo_rl1.fit(x1, y)
modelo_rl2 = LinearRegression()
modelo_rl2.fit(x2, y)
modelo_rl3 = LinearRegression()
modelo_rl3.fit(x3, y)
```

```python
from sklearn.linear_model import LinearRegression

# Create modelo_rl1
modelo_rl1 = LinearRegression()
modelo_rl1.fit(x1, y)

# Create modelo_rl2
modelo_rl2 = LinearRegression()
modelo_rl2.fit(x2, y)

# Create modelo_rl3
modelo_rl3 = LinearRegression()
modelo_rl3.fit(x3, y)
```

```python
# Print R-squared for modelo_rl1
r2_modelo_rl1 = modelo_rl1.score(x1, y)
print("R-squared for modelo_rl1:", r2_modelo_rl1)

# Print R-squared for modelo_rl2
r2_modelo_rl2 = modelo_rl2.score(x2, y)
print("R-squared for modelo_rl2:", r2_modelo_rl2)

# Print R-squared for modelo_rl3
r2_modelo_rl3 = modelo_rl3.score(x3, y)
print("R-squared for modelo_rl3:", r2_modelo_rl3)
```

```python
# Calculate R-squared values
r2_modelo_rl1 = modelo_rl1.score(x1, y)
r2_modelo_rl2 = modelo_rl2.score(x2, y)
r2_modelo_rl3 = modelo_rl3.score(x3, y)

# Find the model with the highest R-squared value
max_r2 = max(r2_modelo_rl1, r2_modelo_rl2, r2_modelo_rl3)
if max_r2 == r2_modelo_rl1:
    print("Modelo modelo_rl1 tiene la mayor bondad de ajuste.")
elif max_r2 == r2_modelo_rl2:
    print("Modelo modelo_rl2 tiene la mayor bondad de ajuste.")
else:
    print("Modelo modelo_rl3 tiene la mayor bondad de ajuste.")
```

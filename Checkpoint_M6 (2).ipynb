```python
import pandas as pd
import statsmodels.api as sm

# Assuming 'df' is your DataFrame

# Define features and target
X = data.drop('Sales', axis=1)  # Include all other columns as features
y = data['Sales']

# Add a constant to the features
X = sm.add_constant(X)

# Create the OLS model and fit it
model = sm.OLS(y, X).fit()

# Print the summary to get R² and other statistics
print(model.summary())
```
